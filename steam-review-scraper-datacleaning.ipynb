# Steam Game Review Scraper Data cleaning

Do data cleaning such as 
- date formating
- review text formating (and review length recalculation)
- play hours formating

## Read the results (csv)

```python
import pandas as pd
from pathlib import Path

# read from csv
file_path = Path("steam-reviews-scrapper-result-20231011.csv")
df = pd.read_csv(file_path)

df
```

```python
print(df.iloc[1]['ReviewText'])
```

```python
df[df['PlayHours'].str.contains('on record') == False]
```

```python
df2 = df[df['PlayHours'].str.contains('on record') == True]

df2
```

```python
# format DatePosted
from datetime import datetime

# classify have year or not
# have year -> before 20230101
# no year -> assume 2023

# Format 'DatePosted' column to datetime objects
# First formatting automatically.
df2['DatePosted2'] = pd.to_datetime(df2['DatePosted'].str.replace('Posted: ', '', regex=False), errors='coerce', infer_datetime_format=True)

df2
```

```python
for index, row in df2.iterrows():
    if pd.notnull(row['DatePosted2']):
        continue

    # then further formatting with specify date format
    try:
        df2.at[index, 'DatePosted2'] = datetime.strptime(str(row['DatePosted']).replace('Posted: ', ''), '%d %B').replace(year=2023)
    except ValueError:
        try:
            df2.at[index, 'DatePosted2'] = datetime.strptime(str(row['DatePosted']).replace('Posted: ', ''), '%B %d').replace(year=2023)
        except ValueError:
            print(f"Row {index}; Cannot convert to datetime")
            print(row)

            print()

df2
```

```python
df2.info()
```

```python
# replace DatePosted2 -> DatePosted

df2_cleaned = df2.drop(columns=['DatePosted'])
df2_cleaned.rename(columns={'DatePosted2': 'DatePosted'}, inplace=True)

df2_cleaned
```

```python
# clean columns "PlayHours"

import locale

def convert(s, thousands_delim = ',', locale_abbr = 'en_US.UTF-8'):
    # for formatting numbers with comma
    locale.setlocale(locale.LC_ALL, locale_abbr)
    try:
        number = locale.atof("".join(s.split(thousands_delim)))
    except ValueError:
        number = None

    return number

df2_cleaned['PlayHours'] = df2_cleaned['PlayHours'].str.replace('hrs on record', '')
df2_cleaned['PlayHours'] = df2_cleaned['PlayHours'].apply(lambda x: convert(x))

df2_cleaned
```

```python
# convert review to number
# 0 -> not recommended
# 1 -> recommended

df2_cleaned['Recommended'] = df2_cleaned['Review'].apply(lambda x: 0 if 'not' in x.lower() else 1)

df2_cleaned
```

```python
df2_cleaned.drop(columns=['Review'], inplace=True)

df2_cleaned
```

```python
# reorder columns

df2_cleaned = df2_cleaned[
    ['GameId', 'SteamId', 'ProfileURL', 'ReviewText', 'Recommended', 'ReviewLength(Chars)', 'PlayHours', 'DatePosted']
]

df2_cleaned
```

```python
import numpy as np

np.array(df2_cleaned['ReviewText'].str.contains('EARLY ACCESS REVIEW')).nonzero()[0]
```

Here we create a new column to indicate Early Review Comments.

And remove the keyword 'EARLY ACCESS REVIEW' from the review text

```python
df2_cleaned['EarlyAccessReview'] = np.where(
    df2_cleaned['ReviewText'].str.contains('EARLY ACCESS REVIEW'), True, False
)

df2_cleaned
```

```python
# then we strip the 'EARLY ACCESS REVIEW'

df2_cleaned['ReviewText'] = df2_cleaned['ReviewText'].str.replace('EARLY ACCESS REVIEW', '')
df2_cleaned['ReviewText'] = df2_cleaned['ReviewText'].str.strip('\n')

df2_cleaned
```

```python
# and recalculate the length of reviews
# the method is copied from steam-review-scraper.ipynb

df2_cleaned['ReviewLength(Chars)'] = df2_cleaned.apply(lambda x: len(str(x['ReviewText']).replace(' ', '')), axis=1)

df2_cleaned
```

```python
# last 8 characters are yyyymmdd
df2_cleaned.to_pickle(
    file_path.stem[:-8] + 'cleaned-' + file_path.stem[-8:] + '.pkl'
)
```
