```python
# how about only keep some punctuations?
keep_punct = [".", ",", "!", "?", ":"]
```

```python
def remove_punctuations(text):
    final = "".join(c for c in text if c not in set(string.punctuation))
    return final

l4 = l3.copy()

l4['review_text'] = l4['review_text'].apply(remove_punctuations)

l4
```

```python
# only keep letters (A-Z, a-z)

def remove_non_letters(text):
    final = "".join(c for c in text if c.isalpha() or c == " ")
    return final

l5 = l4.copy()

l5['review_text'] = l5['review_text'].apply(remove_non_letters)

l5
```

```python
# remove numbers using regex

import re

def remove_numbers(text):
    final = re.sub(r'\d+', '', text)
    return final
```

```python
# keep only A-Z, a-z, and spaces with regex

import re

def remove_non_letters(text):
    final = re.sub(r'[^a-zA-Z ]+', ' ', text)
    return final

l6 = l5.copy()

l6['review_text'] = l6['review_text'].apply(remove_non_letters)

l6
```

```python
def cleaning_df(df,review):
    '''apply all cleaning functions to a dataframe column
    '''
    df[review] = df[review].apply(str_cleaning_functions.clean)
    df[review] = df[review].apply(str_cleaning_functions.deEmojify)
    df[review] = df[review].apply(str_cleaning_functions.remove_non_letters)
    df[review] = df[review].str.lower()
    # df[review] = df[review].apply(remove_num)
    # df[review] = df[review].apply(remove_symbols)
    # df[review] = df[review].apply(remove_punctuation)
    df[review] = df[review].apply(str_cleaning_functions.unify_whitespaces)
    # df[review] = df[review].apply(str_cleaning_functions.remove_stopword)
    # df[review] = df[review].apply(str_cleaning_functions.unify_whitespaces)
    # df[review] = df[review].apply(str_cleaning_functions.stemming)
    # df[review] = df[review].apply(str_cleaning_functions.unify_whitespaces)
```
