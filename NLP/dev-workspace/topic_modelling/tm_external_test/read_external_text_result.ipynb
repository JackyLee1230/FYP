```python
eval_results_external_folder_path = Path(f'../eval_results_external/{game_name}')

print(eval_results_external_folder_path)

if not eval_results_external_folder_path.exists():
    eval_results_external_folder_path.mkdir(parents=True)
```

Read LDA result

```python
# load the LDA model
%autoreload 2
from dataset_loader import GENRES

genre = GENRES.ACTION
training_datetime = datetime(2024, 2, 27, 9, 18, 50)
N_topics = 10

lda_model_folder = Path(f'../lda_dev/category_{str(genre)}_unique_review_text')
lda_model_folder = lda_model_folder.joinpath(
    Path(f'lda_multicore_genre_{str(genre)}_grid_search_{training_datetime.strftime("%Y%m%d_%H%M%S")}')
)
lda_model_folder = lda_model_folder.joinpath(
    Path(f'lda_multicore_lda_num_topics_{N_topics}')
)
```

```python
eval_results_external_folder_path_lda = eval_results_external_folder_path.joinpath(
    *lda_model_folder.parts[2:]
)

eval_results_external_folder_path
```

```python
# read the top 10 topic frequency

top_10_topic_freq_lda = pd.read_pickle(
    eval_results_external_folder_path_lda.joinpath('df_eval_topic_freq.pkl')
)

top_10_topic_freq_lda
```

```python
# read top 10 repr docs

top_10_repr_docs_lda = pd.read_pickle(
    eval_results_external_folder_path_lda.joinpath('df_eval_top_10.pkl')
)

top_10_repr_docs_lda
```

```python
# read the llm generation result


with open(eval_results_external_folder_path_lda.joinpath(
    'llm_generation_result.json'), 'r') as f:
    llm_generation_result_lda = json.load(f)


llm_generation_result_lda
```

```python
# load top 10 keywords

with open(eval_results_external_folder_path_lda.joinpath(
    'top_10_keywords.json'), 'r') as f:
    top_10_keywords_lda = json.load(f)

top_10_keywords_lda
```

```python
# create a dataframe that shows the topic frequency, top 10 keywords and all possible topic names

df_new_topic_labels_lda = pd.DataFrame(llm_generation_result_lda['new_topic_labels'])
df_new_topic_labels_lda = df_new_topic_labels_lda.T
df_new_topic_labels_lda.reset_index(inplace=True)
df_new_topic_labels_lda.rename(columns={'index': 'topic_id'}, inplace=True)
# convert topic_id to int
df_new_topic_labels_lda['topic_id'] = df_new_topic_labels_lda['topic_id'].astype(int)

df_new_topic_labels_lda
```

```python
df_eval_lda = pd.merge(
    top_10_topic_freq_lda, 
    df_new_topic_labels_lda, 
    on='topic_id'
)

df_eval_lda
```
