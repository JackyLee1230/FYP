```python
# temporary script to group pickles into a single file

# created on: 09-10-2023

from pathlib import Path
import os

import pandas as pd

dataset_folder = Path('dataset_cleaned_heartless_sampled_20231129').resolve()

for root, dirs, files in os.walk(dataset_folder):
    target_pkls = list(map(lambda f: Path(root, f), files))
    target_pkls = [p for p in target_pkls if p.suffix == '.pkl' and p.stem.find('cleaned_20231211') != -1]

    break   # only one-level

print(len(target_pkls))
print()
for pkl in target_pkls:
    print(pkl)
```

```python
# join them and save a new pickle



df = pd.concat([pd.read_pickle(p) for p in target_pkls])
    
df
```

```python
df.reset_index(inplace=True, drop=True)

df
```

```python
# save

from datetime import datetime


save_path = Path(dataset_folder, f'dataset_cleaned_heartless_sampled_20231129_joincleanedchunks_{datetime.now().strftime("%Y%m%d")}.pkl')

df.to_pickle(save_path)
print(f"Saved to: {save_path}")
```
